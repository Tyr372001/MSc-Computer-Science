## Program-1: **Find the Sum of First 'n' Natural Numbers**





```python
# Finding the sum of First 'n' Natural Numbers

sum = 0
n = int(input("Enter the value of n: "))


while(n>0):
  sum = sum + n
  n = n - 1

print("\n")

print("The sum of first ", n," natural numbers is: ",sum)

```



## Program-2: **Print Even Numbers Between 'm' to 'n'**



```python
# Printing even numbers between 'm' and 'n'

m = int(input("Enter the value of m: "))
n = int(input("Enter the value of n: "))

print("The even numbers between ",m," and ",n," are: ")
print("\n")


while(m<=n):
  if m%2 == 0:
    print(m)
  m = m + 1

```

## Program-3 **Print prime numbers between 'm' to 'n'**

```python
# Asking inputs from the user

m = int(input("Enter the value of m: "))
n = int(input("Enter the value of n: "))


print("The prime numbers between ",m," and ",n," are: ")
print("\n")

# A nested while-loop to compute and determine prime numbers between m and n

while m <= n:
    count = 0
    i = 1

    while i <= m:
        if m % i == 0:
            count += 1
        i += 1

    if count == 2:
        print(m)
    m += 1

```

## Program-4: **Display the factorial of a given number**

```python

# asking for the number from the user
n = int(input("Enter the number: "))
temp = n
fact = 1
while(n>0):
  fact = fact * n
  n = n - 1

print("The factorial of {} is {}".format(temp,fact))
```

## Program-6: **Check the given number is Armstrong Number or not**

```python
# Asking for number from the user
number = int(input("Enter the number: "))

sum = 0
temp = number

while temp > 0:
  digit = temp % 10
  sum += digit ** 3
  temp //= 10

if number == sum:
  print(number,"is an Armstrong number")
else:
  print(number,"is not an Armstrong number")
```

## Program 7: **Find the pass-percentage of a class. A teacher is entering the marks of students. A student passes a course if the marks are atleast 40 (out of 100). The Teacher wants to know the percentage of the student passed**

$$ \text{Pass Percentage} = \frac{ \times 100}{} \\ $$

$$   \text{Number of Student Passed} \\ \\  \rightarrow \text{Total Number of Student in the Class} $$

```python
pass_count = 0
total_count = 0

while True:
  marks = int(input("Enter the Marks of the Student:   "))
  print()
  print()


  if marks < 40:
    print("Sorry!! The student FAILED!!")
    total_count += 1
    print()
    print()


  else:
    pass_count = pass_count + 1
    total_count += 1
    print("Congratulations!! The student PASSED!!")
    print()
    print()


  flag = input("Do you want to continue? 1 == YES otherwise Exit with showing the pass percentage:   ")
  print()
  print()


  if flag != "1":
    pass_percentage = (pass_count / total_count) * 100
    print("The pass percentage of the class is: {:.2f}% ".format(pass_percentage))
    break
```

## Program-12 **Display the following pattern of *s (Pascal's Triangle)**


pattern pascal triangle.jpg

```python
# Number of rows
n = 5

# Loop through each row
for i in range(n):
    # Print leading spaces
    for j in range(n - i - 1):
        print(" ", end="")

    # Print Pascal's Triangle values using *
    num = 1
    for j in range(i + 1):
        print("*", end=" ")
        num = num * (i - j) // (j + 1)

    # Move to the next line
    print()
    print()
```
