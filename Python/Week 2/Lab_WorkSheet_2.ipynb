```python
# Asking a number from the user
m = int(input("Enter the value of m: "))
n = int(input("Enter the value of n: "))

print()
print("The Perfect numbers between", m, "and", n, "are:")
print("\n")

for i in range(m, n):
    divisor_sum = 0
    for j in range(1, i):  # Start from 1 to i-1 to find divisors
        if (i % j) == 0:
            divisor_sum += j  # Accumulate the sum of divisors

    if i == divisor_sum:
        print(i)  # Print the perfect number
```

```python
# Function to calculate the sum of divisors of a number n
def sum_of_divisors(n):
    total = 1  # 1 is always a divisor
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            total += i
            if i != n // i:
                total += n // i
    return total

# Find and print all amicable pairs between 1 and 100000
limit = 100000

for a in range(2, limit + 1):
    b = sum_of_divisors(a)  # sum of divisors of a
    if b > a and b <= limit:  # Ensure b is greater than a to avoid repetitions
        if sum_of_divisors(b) == a:  # Check if the sum of divisors of b equals a
            print(f"Amicable Pair: ({a}, {b})")
```

```python
# Asking integer from the user

n = int(input("Enter a decimal number: "))

binary_representation = ""

# Edge case for 0
if n == 0:
    binary_representation = "0"
else:
    temp = n
    while temp > 0:
        remainder = temp % 2  # Get the remainder when divided by 2
        binary_representation = str(remainder) + binary_representation  # Append remainder to the binary string
        temp = temp // 2  # Divide by 2 and continue

print("Binary representation of {} is {}".format(n,binary_representation))
```

```python
# Asking from the user the values of X and n
import math

def factorial(n):
  factorial = 0
  if(n == 0):
    return 1
  else:
    for i in range(1,n):
      factorial = factorial * i;
    return factorial


x = int(input("Enter the value of x: "))
n = int(input("Enter the value of n: "))

result = 0

for i in range(1, n+1):
  result = result + (i + factorial(math.floor(x/i)))

print()

print("The total Summation is : {}".format(result))
```
