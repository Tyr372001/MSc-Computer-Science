# 1. Write a function that checks whether a number is in a given range or not(Inclusive of High and Low)

```python
def is_in_range(number, low, high):
    """
    Check if a number is within a given range (inclusive).

    Parameters:
    number (int or float): The number to check.
    low (int or float): The lower bound of the range.
    high (int or float): The upper bound of the range.

    Returns:
    bool: True if the number is in the range [low, high], False otherwise.
    """
    return low <= number <= high


number = float(input("Enter the Number: "))
high = float(input("Enter the High: "))
low = float(input("Enter the Low: "))

if is_in_range(number, low, high):
    print(f"{number} is in the range between {low} and {high}.")
else:
    print(f"{number} is not in the range between {low} and {high}.")
```

# 2. Write a Python function that accepts a string and calculates the number of uppercase letters and lower case letters.

```python
def count_case_letters(input_string):
    """
    Count the number of uppercase and lowercase letters in a string.

    Parameters:
    input_string (str): The string to analyze.

    Returns:
    tuple: A tuple containing the count of uppercase letters and lowercase letters.
    """
    upper_count = 0
    lower_count = 0

    for char in input_string:
        if char.isupper():
            upper_count += 1
        elif char.islower():
            lower_count += 1

    return upper_count, lower_count

string = input("Enter the String: ")
upper_count, lower_count = count_case_letters(string)

print(f"Number of Uppercase Letters: {upper_count}")

print(f"Number of Lowercase Letters: {lower_count}")
```

# 3. Write a Python function to multiply all arguments

```python
def multiply_all(*args):
    """
    Multiply all arguments provided to the function.

    Parameters:
    *args: A variable number of numeric arguments.

    Returns:
    int or float: The product of all arguments. Returns 1 if no arguments are provided.
    """
    product = 1
    for number in args:
        product *= number
    return product

result_1 = multiply_all(1, 2, 3, 4)
print(result_1)

result_2 = multiply_all(5, 6, 7)
print(result_2)

result_3 = multiply_all(3,4)
print(result_3)

```

# 4. Write a Python function that checks whether a word or phrase is palindrome or not.

```python
def is_palindrome(input_string):
    """
    Check if a word or phrase is a palindrome.

    Parameters:
    input_string (str): The string to check.

    Returns:
    bool: True if the string is a palindrome, False otherwise.
    """
    # Normalize the string: remove spaces and convert to lowercase
    normalized_string = ''.join(char.lower() for char in input_string if char.isalnum())

    # Check if the normalized string is equal to its reverse
    return normalized_string == normalized_string[::-1]

string = input("Enter the String: ")

if is_palindrome(string):
    print(f"'{string}' is a palindrome.")
else:
    print(f"'{string}' is not a palindrome.")
```

# 5. Write a Python function to check whether a string is pangram or not. (Assume the string passed in does not have any punctuation)

```python
def is_pangram(input_string):
    """
    Check if a string is a pangram.

    Parameters:
    input_string (str): The string to check.

    Returns:
    bool: True if the string is a pangram, False otherwise.
    """
    # Create a set of all letters in the English alphabet
    alphabet_set = set('abcdefghijklmnopqrstuvwxyz')

    # Convert the input string to lowercase and create a set of its characters
    input_set = set(input_string.lower())

    # Check if all alphabet letters are in the input set
    return alphabet_set.issubset(input_set)

string = input("Enter the String: ")

if is_pangram(string):
    print(f"'{string}' is a pangram.")
else:
    print(f"'{string}' is not a pangram.")
```

# 6. Implement a function to calculate the factorial of a number.

```python
def factorial_iterative(n):
    """
    Calculate the factorial of a non-negative integer using an iterative approach.

    Parameters:
    n (int): The non-negative integer for which to calculate the factorial.

    Returns:
    int: The factorial of the number n.
    """
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers.")

    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

number = int(input("Enter the Number: "))

result = factorial_iterative(number)
print(f"The factorial of {number} is {result}.")
```

# 7.  Python Program to Find HCF or GCD through function

```python
def gcd(a, b):
    """
    Calculate the GCD (HCF) of two numbers using the Euclidean algorithm.

    Parameters:
    a (int): The first number.
    b (int): The second number.

    Returns:
    int: The GCD of the two numbers.
    """
    while b:
        a, b = b, a % b
    return abs(a)  # Return the absolute value to handle negative inputs

num1 = int(input("Enter the First Number: "))
num2 = int(input("Enter the Second Number: "))

result = gcd(num1, num2)
print(f"The GCD of {num1} and {num2} is {result}.")
```

# 8. Check the number is prime or not using python function

```python
def is_prime(n):
    """
    Check if a number is prime.

    Parameters:
    n (int): The number to check.

    Returns:
    bool: True if the number is prime, False otherwise.
    """
    count = 0
    for i in range(1, n + 1):
        if n % i == 0:
            count += 1
    return count == 2

number = int(input("Enter the Number: "))

if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

# 9.  Write a program that takes a string as input and:
## i. Print the length of the string
## ii. Print the first and last character of the string
## iii. Prints the middle character (if the length is odd) or the two middle characters (if the length is even)

```python
def analyze_string(input_string):
    """
    Analyze the given string and print its length, first and last characters,
    and middle character(s).

    Parameters:
    input_string (str): The string to analyze.
    """
    # i. Print the length of the string
    length = len(input_string)
    print(f"Length of the string: {length}")

    # ii. Print the first and last character of the string
    if length > 0:
        first_char = input_string[0]
        last_char = input_string[-1]
        print(f"First character: {first_char}")
        print(f"Last character: {last_char}")

        # iii. Print the middle character(s)
        if length % 2 == 1:
            middle_index = length // 2
            middle_char = input_string[middle_index]
            print(f"Middle character: {middle_char}")
        else:
            middle_index1 = (length // 2) - 1
            middle_index2 = length // 2
            middle_chars = input_string[middle_index1:middle_index2 + 1]
            print(f"Middle characters: {middle_chars}")
    else:
        print("The string is empty.")

# Example Usage
input_str = input("Enter a string: ")
analyze_string(input_str)
```

# 11. Write a program to count the number of vowels and consonants in a string

```python
def count_vowels_and_consonants(input_string):
    """
    Count the number of vowels and consonants in a string.

    Parameters:
    input_string (str): The string to analyze.

    Returns:
    tuple: A tuple containing the count of vowels and consonants.
    """
    # Define vowels
    vowels = "aeiouAEIOU"

    # Initialize counts
    vowel_count = 0
    consonant_count = 0

    # Iterate through each character in the string
    for char in input_string:
        if char.isalpha():  # Check if the character is an alphabetic letter
            if char in vowels:
                vowel_count += 1  # Increment vowel count
            else:
                consonant_count += 1  # Increment consonant count

    return vowel_count, consonant_count


input_str = input("Enter a string: ")
vowels, consonants = count_vowels_and_consonants(input_str)
print(f"Number of vowels: {vowels}")
print(f"Number of consonants: {consonants}")
```

# 12.  Write a program to accept sentence and find the longest word in a sentence

```python
def find_longest_word(sentence):
    """
    Find the longest word in a given sentence.

    Parameters:
    sentence (str): The sentence to analyze.

    Returns:
    str: The longest word in the sentence.
    """
    # Split the sentence into words
    words = sentence.split()

    # Initialize variables to track the longest word
    longest_word = ""

    # Iterate through each word in the list
    for word in words:
        # Update longest_word if current word is longer
        if len(word) > len(longest_word):
            longest_word = word

    return longest_word

# Example Usage
input_sentence = input("Enter a sentence: ")
longest_word = find_longest_word(input_sentence)
print(f"The longest word is: '{longest_word}'")
```

# 13. Count the number of words in a given sentence

```python
def count_words(sentence):
    """
    Count the number of words in a given sentence.

    Parameters:
    sentence (str): The sentence to analyze.

    Returns:
    int: The number of words in the sentence.
    """
    # Split the sentence into words using whitespace as the delimiter
    words = sentence.split()

    # Return the number of words
    return len(words)

# Example Usage
input_sentence = input("Enter a sentence: ")
word_count = count_words(input_sentence)
print(f"The number of words in the sentence is: {word_count}")
```

#. 14.  Write a program to remove duplicate characters in a string.


```python
def remove_duplicates(input_string):
    """
    Remove duplicate characters from a string while preserving the order.

    Parameters:
    input_string (str): The string from which to remove duplicates.

    Returns:
    str: A string with duplicate characters removed.
    """
    # Initialize an empty string to store the result
    result = ""

    # Iterate through each character in the input string
    for char in input_string:
        # Add the character to the result if it's not already included
        if char not in result:
            result += char

    return result

# Example Usage
input_str = input("Enter a string: ")
output_str = remove_duplicates(input_str)
print(f"String after removing duplicates: '{output_str}'")
```

# 15.  Write a program to remove all punctuation marks from a string.

```python
import string

def remove_punctuation(input_string):
    """
    Remove all punctuation marks from a string.

    Parameters:
    input_string (str): The string from which to remove punctuation.

    Returns:
    str: A string with all punctuation marks removed.
    """
    # Create a translation table that maps each punctuation character to None
    translator = str.maketrans('', '', string.punctuation)

    # Use the translate method to remove punctuation
    return input_string.translate(translator)

# Example Usage
input_str = input("Enter a string: ")
output_str = remove_punctuation(input_str)
print(f"String after removing punctuation: '{output_str}'")
```
